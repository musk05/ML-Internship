# QUES - What is the average number of words in movie titles between the year 2000-2005?

```python
movies= df[(df['release_year'] >= 2000) & (df['release_year'] <= 2005)]
movies
```

```python
movies['original_title'].shape
```

```python
  count= 0 
for title in movies['original_title']:
    words = title.split(' ')
    count = count +len(words)
total_entries = len(movies['original_title'])
meanwords = count/total_entries
print("The average number of words in movie titles between the year 2000-2005 is : (%d) "%(meanwords) + str(meanwords) )
```
